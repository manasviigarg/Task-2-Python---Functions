```python
#Q2 - Write a function to extract first word of s string (with many words separated by space)
def extfirstword():
    s = input('Enter a string - ')
    l = len(s)
    print('Length of the String is', l)
    l1 = []
    for i in range(0, l+1):
        if s[i] != ' ':
            l1.append(s[i])
        else :
            break
    s1 = ''.join(l1)
    print(s1)
extfirstword()
```

```python
#Q3 - Extract the first word from every string from a list of strings by using map function
l1 = ['Don\'t underestimate the allure of the darkness. Even the purest hearts are drawn to it.',
      'This family makes me want to murder people.',
      'He\'s your first love. I intend to be your last. However long it takes.',
      'Family can be more than just those with whom we share blood. We can choose.',
      'Always and forever is not just something you weasel out of, brother.']
def fword(string):
    return string.split()[0]
res = list(map(fword,l1))
print(res)
```

```python
#Q4 - Write a function to return a list of prime factors of a given number
def primefactor():
    n = int(input('Enter a number '))
    l1 = []
    for i in range(1, (n*n)+2):
        if n % i == 0:
            l1.append(i)
        else: 
            continue
    print('Factors of the number are', l1)
    length = len(l1)
    l2 = []
    for i in range(0, length):
        if l1[i] > 1:
            for j in range(2, l1[i]):
                if l1[i] % j == 0:
                    break
            else :
                l2.append(l1[i])
    print('Prime Factors are',l2)
primefactor()
```

```python
#Q.5 - Write a function that finds 2nd largest among 4 numbers (Repetitions are allowed, without sorting)
def seclar():
    l1 = []
    for i in range(0, 4):
        x = int(input('Enter the element '))
        l1.append(x)
    print('List is ', l1)
    x = max(l1)
    l1.remove(x)
    for i in range(0,3):
        if x == l1[i]:
            l1.remove(x)
            break
    y = max(l1)
    print('2 largest number is', y)
seclar()
```
